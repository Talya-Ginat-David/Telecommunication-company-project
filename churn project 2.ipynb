```python
##replace blank string with NaN 
churn = churn.replace(r'^\s*$', np.nan, regex=True)
churn
```

```python
##now we can see that there are 11 missing values in total charges column.
churn.isnull().sum()
```

```python
##to find NaN values
churn.isna().mean()
```

```python
#changing totalcharhes from object(string) to float
churn['totalcharges'] = churn['totalcharges'].astype(float)
#filling total charges with monthlycharges*tenure
churn['totalcharges'] = churn.apply(\
                lambda row: row['tenure'] * row['monthlycharges'] if pd.isna(row['totalcharges']) else row['totalcharges'], axis=1)
churn.head(10)
```

```python
#set customerid to index
churn = churn.set_index('customerid')
churn
```

```python
churn.columns
```

```python
## no missing values :)
churn.isna().mean()
```

```python
##changing string (object) to int 
churn['churn'] = churn['churn'].apply(lambda label: 0 if label == 'No' else 1)
churn['phoneservice'] = churn['phoneservice'].apply(lambda label: 0 if label == 'No' else 1)
churn['internetservice'] = churn['internetservice'].apply(lambda label: 0 if label == 'No' else 1)
churn['streamingtv'] = churn['streamingtv'].apply(lambda label: 0 if label == 'No' else 1)
churn['partner'] = churn['partner'].apply(lambda label: 0 if label == 'No' else 1)
churn['dependents'] = churn['dependents'].apply(lambda label: 0 if label == 'No' else 1)
churn['paperlessbilling'] = churn['paperlessbilling'].apply(lambda label: 0 if label == 'No' else 1)
churn['multiplelines'] = churn['multiplelines'].apply(lambda label: 1 if label == 'Yes' else 0)
churn['onlinesecurity'] = churn['onlinesecurity'].apply(lambda label: 1 if label == 'Yes' else 0)
churn['onlinebackup'] = churn['onlinebackup'].apply(lambda label: 1 if label == 'Yes' else 0)
churn['deviceprotection'] = churn['deviceprotection'].apply(lambda label: 1 if label == 'Yes' else 0)
churn['techsupport'] = churn['techsupport'].apply(lambda label: 1 if label == 'Yes' else 0)
churn['streamingmovies'] = churn['streamingmovies'].apply(lambda label: 1 if label == 'Yes' else 0)
churn['gender'] = churn['gender'].apply(lambda label: 1 if label == 'Female' else 0)
churn
```

```python
def plot_feature_by_label_histogram(df, feature_col, label_col):
    df_counts = (churn.groupby([label_col])[feature_col]
                     .value_counts(normalize=True)
                     .rename('percentage')
                     .mul(100)
                     .reset_index())
    p = sns.barplot(x=feature_col, y="percentage", hue=label_col , data=df_counts)
```

```python
from sklearn.model_selection import train_test_split

label = 'churn'
cust = 'customerid' 
features = [c for c in churn.columns if c not in ['gender', label]]



x_train, x_test, y_train, y_test = train_test_split(churn[features], churn[label],
                                                    stratify=churn[label], 
                                                    test_size=0.20)

```

```python
from sklearn.tree import DecisionTreeClassifier 

clf = DecisionTreeClassifier(max_depth=3)


clf.fit(x_train, y_train) 

y_test_pred_DecisionTree = clf.predict(x_test) 
y_test_pred_DecisionTree
output= x_test.copy()
output[label] = y_test
output['churn_predicted_by_model']= y_test_pred_DecisionTree
output= output[['churn_predicted_by_model', label]].rename(columns={label:'churn_what_actualy_happened'}).reset_index()

output.to_csv('my_DecisionTree_Prediction.csv', index=False) 
output
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from IPython.display import SVG
from graphviz import Source
from IPython.display import display


def plot_tree(tree, features, labels):
    graph = Source(export_graphviz(tree, feature_names=features, class_names=labels, filled = True))
    display(SVG(graph.pipe(format='svg')))

plot_tree(clf, x_train.columns, ['stayed','churn'])
```

```python
def create_matrix_figure(matrix_df, figsize):
       
        fig = plt.figure(figsize=figsize)
        plt.clf()
        plt.grid(b=None)
        ax = fig.add_subplot(111)
        ax.set_aspect(1)
        res = ax.imshow(matrix_df.values, cmap=plt.cm.jet, interpolation='nearest', vmax=1.0, vmin=0.0, aspect='equal')
        width, height = matrix_df.shape
        for x in range(width):
            for y in range(height):
                ax.annotate(float('%.2f' % matrix_df.iloc[x][y]), xy=(y, x), \
                            horizontalalignment='center', verticalalignment='center')

        ax.set_xticks(range(width))
        ax.set_yticks(range(height))
        ax.set_xticklabels(matrix_df.columns.tolist())
        ax.set_yticklabels(matrix_df.columns.tolist())
        ax.xaxis.tick_top()
        return fig
```
